#Bivariate analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('train.csv')
```

```python
df
```

```python
cols = ['LotFrontage', 'LotArea',  'MasVnrArea', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', 
       '2ndFlrSF', 'BsmtHalfBath', 'GarageYrBlt', 'GarageArea', 'PoolArea', 'MiscVal', 'SalePrice']
```

```python
df = df[cols]
```

```python
sns.scatterplot(data=df, x= 'GarageArea', y='SalePrice')
```

```python
sns.scatterplot(data=df, x= 'TotalBsmtSF', y='SalePrice')
```

```python
plt.figure(dpi=140)
sns.pairplot(df)
plt.show()
```
