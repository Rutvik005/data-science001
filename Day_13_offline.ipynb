#Draw  PDF and CDF

```
counts, bin_edges = np.histogram(df_num['SalePrice'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))

cdf = np.cumsum(pdf)
a , =plt.plot(bin_edges[1:],pdf);
b , =plt.plot(bin_edges[1:], cdf)
plt.ylabel("Probablity")
plt.legend([a,b],['PDF','CDF'])
plt.show()
```

```
counts, bin_edges = np.histogram(df_num['GarageArea'], bins=10, density = True)

pdf = counts/(sum(counts))

cdf = np.cumsum(pdf)
a , =plt.plot(bin_edges[1:],pdf);
b , =plt.plot(bin_edges[1:], cdf)
plt.ylabel("Probablity")
plt.legend([a,b],['PDF','CDF'])
plt.show()
```
